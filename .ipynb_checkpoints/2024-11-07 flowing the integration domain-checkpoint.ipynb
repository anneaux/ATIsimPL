### config

```julia
const λ = 800 # nm
const Intensity0 = 4e14 # W/cm2
const Ip = IpAU(15.7596) # Argon gas target

const ω = 2 * π * LAU * c / λ
const TC = 2*π/ω;
```

# field configurations

(These are just some arbitrary test configurations)

```julia
beam1 = let # a simple monochromatic field
    beam = BeamMono(Intensity = Intensity0, 
    lambda=λ, epsilon=0.,
    phi = 0.)
end
```

```julia
beam2 = let # co-linear two-colour field with mixing angle θ, "tunnelling without barrier"-scenario
    θ = 0.#π/4
    beam = BeamBETC(Intensity1 = (cos(θ))^2 * Intensity0, 
    Intensity2 = (sin(θ))^2 * Intensity0,
    lambda=λ, r=1,s=2, epsilon=0.,
    phi = +π/2)
end
```

```julia
beam3 = let # fully 2ω field
    θ = π/2
    beam = BeamBETC(Intensity1 = (cos(θ))^2 * Intensity0, 
    Intensity2 = (sin(θ))^2 * Intensity0,
    lambda=λ,r=1,s=2,epsilon=0.,
    phi = 0.)
end
```

```julia
beam4 = let # just a random other weird field
    θ = 0.3 #π/6
    beam = BeamBETC(Intensity1 = (cos(θ))^2 * Intensity0, 
    Intensity2 = (sin(θ))^2 * Intensity0,
    lambda=λ,r=1,s=2,epsilon=0.,
    phi = +π/3)
end
```

## plot the field shape over time

```julia
let
    beam = beam1

    Nt = 150
    times = range(-TC, stop = TC, length = Nt)    
    
    efield = [E(beam)(t)[1] for t in times] # only the [1] component (x)

    f = Figure()
    ax = Axis(f[1, 1],
        xlabel = "time / opt. cycle",
        ylabel = "el. field",
    ) 
    
    lines!(ax, times, efield)

    f
end
```

# action landscape with flowed integration domain

```julia
let
    beam = beam1
    momentum = [0.,0.] # wants to be a Vector
    Nsteps = 30 # number of flow steps
    
    ####################
    
    ### define your phase functions, as functions of 't' only
    S(t) = action(beam, momentum, t)
    S_drv(t) = action_drv(beam, momentum, t)
    
    tmin = 0. + 0im # min(t real) + 0.
    tmax = TC + TC/2*im # min(t real) + max(t imag)
    
    ### get saddle points
    saddles = filter(ts -> real(tmin) < real(ts) < real(tmax), 
        filter(ts->imag(ts)>0, find_saddles_sobol(S_drv, tmin, tmax, 300))
        )
    
    timags = range(imag(tmin), stop = imag(tmax), length = 100)
    treals = range(real(tmin), stop = real(tmax), length = 100)
    
    Svals = [S(tr + im*ti) for tr in treals, ti in timags]
        
    f = Figure() 
    ax = Axis(f[1, 1], aspect=DataAspect(), xlabel = "real time", ylabel = "imaginary time")
    
    cmap = :grayyellow
    contourf!(ax, treals, timags, imag.(Svals), colormap=cmap
        , extendlow = :white, extendhigh = :white
    )
    
    contour!(ax, treals, timags, imag.(-Svals)
        , levels=21, color=:grey, linewidth=0.85)   
    
    contour!(ax, treals, timags, real.(Svals)
        , levels=21, color=:grey, linewidth=0.85)
    
      
    scatter!(ax, reim(saddles)..., color = :black, marker = :x, markersize = 15)

    ### plot thimbles
    points, simplices = get_thimble(S, S_drv, real(tmin), real(tmax),
        Nflow = Nsteps, h_threshold=-500.)
    
    for sim in simplices
        if sim.active
            scatterlines!(ax, [real(points[sim.coord[1]].coord), real(points[sim.coord[2]].coord)],
                [imag(points[sim.coord[1]].coord), imag(points[sim.coord[2]].coord)]
                , markercolor =:darkgreen, linewidth = 2, color = :green, label=false)
        end
    end
    
    f
end
```

# integrand along the deformed contour

```julia
point2ComplexF64(p::Point)=complex((Float64.(p.data))...)
```

```julia
let
    beam = beam1
    momentum = [0.,0.] # wants to be a Vector
    Nsteps = 30
    flowfactor = 1.
    
    
    ### define your phase functions, as functions of 't' only
    S(t) = action(beam, momentum, t)
    S_drv(t) = action_drv(beam, momentum, t)
    
    tmin = 0. + 0im # min(t real) + 0.
    tmax = TC + TC/2*im # min(t real) + max(t imag)
    
    ### get saddle points
    saddles = filter(ts -> real(tmin) < real(ts) < real(tmax), 
        filter(ts->imag(ts)>0, find_saddles_sobol(S_drv, tmin, tmax, 300))
        )
 
    f = Figure(size=(800, 300)) #, fontsize=20)
    ax = Axis(f[1, 1], xlabel = "real part of the contour", ylabel = "integrand along the contour")
    
    points, simplices = get_thimble(S, S_drv, real(tmin), real(tmax),
        Nflow = Nsteps,
        flowstepfactor = flowfactor,
        h_threshold=-500.,
        Δinit = 1.,
        subdividethreshold=0.5)
    
    
    integrand_real = [(sim.active ? 
            ((real(points[sim.coord[1]].coord),
                real(exp(im*S(points[sim.coord[1]].coord))) ),
            ( real(points[sim.coord[2]].coord), 
                real(exp(im*S(points[sim.coord[2]].coord)))   ))
            : nothing) 
        for sim in simplices] 
    
    integrand_imag = [(sim.active ? 
            ((real(points[sim.coord[1]].coord),
                imag(exp(im*S(points[sim.coord[1]].coord))) ),
            ( real(points[sim.coord[2]].coord), 
                imag(exp(im*S(points[sim.coord[2]].coord)))   ))
            : nothing) 
        for sim in simplices]    
    
 
    linesegments!(ax, integrand_real, color = :lightgreen)
    linesegments!(ax, integrand_imag, color = :darkgreen)
    
    text!(ax, "real(exp(i*S))", space=:relative,
        position = (0.01, 0.85), 
        align = (:left, :center),
        fontsize = 18, color = :lightgreen)
    text!(ax, "imag(exp(i*S))", space=:relative,
        position = (0.01, 0.95), 
        align = (:left, :center),
        fontsize = 18, color = :darkgreen)
    
    f
end
```

# movie for the flow of the integration domain

```julia
let
    ### configuration
    beam = beam1
    momentum = [0.,0.] # wants to be a Vector
    ############
    
    
    ### define your phase functions as functions of 't' only
    S(t) = action(beam, momentum, t)
    S_drv(t) = action_drv(beam, momentum, t)
    
    tmin = 0. + 0im # min(t real) + 0.
    tmax = TC + TC/2*im # min(t real) + max(t imag)
    
    ### get saddle points
    saddles = filter(ts -> real(tmin) < real(ts) < real(tmax), 
        filter(ts->imag(ts)>0, find_saddles_sobol(S_drv, tmin, tmax, 300))
        )
    ### initialise the flowed contour (=thimble)
    points, simplices = get_thimble(S, S_drv, real(tmin), real(tmax),
        Nflow = 1,
        flowstepfactor = flowfactor,
        h_threshold=-500.,
        Δinit = 1.,
        subdividethreshold=0.5)
    
    ### real and imaginary part of the integrand exp(iS) along the contour
    integrand_real = Observable([(sim.active ? 
            ((real(points[sim.coord[1]].coord),
                real(exp(im*S(points[sim.coord[1]].coord))) ),
            ( real(points[sim.coord[2]].coord), 
                real(exp(im*S(points[sim.coord[2]].coord)))   ))
            : nothing) 
        for sim in simplices] )
    
    integrand_imag = Observable([(sim.active ? 
            ((real(points[sim.coord[1]].coord),
                imag(exp(im*S(points[sim.coord[1]].coord))) ),
            ( real(points[sim.coord[2]].coord), 
                imag(exp(im*S(points[sim.coord[2]].coord)))   ))
            : nothing) 
        for sim in simplices] )
    
    ### making the figure
    f = Figure()
    
    ### top panel: integrand
    ax1 = Axis(f[1, 1],
        xlabel = "real part of the contour", ylabel = "integrand"
        , yticklabelspace = 30., xticks = [0,50,100])
     xlims!(ax1, (real(tmin), real(tmax)))
    
    linesegments!(ax1, integrand_real, color = :lightgreen)
    linesegments!(ax1, integrand_imag, color = :darkgreen)
    
    text!(ax1, "real(exp(i*S))", space=:relative,
        position = (0.01, 0.75), align = (:left, :center), fontsize = 16, color = :lightgreen)
    text!(ax1, "imag(exp(i*S))", space=:relative,
        position = (0.01, 0.9), align = (:left, :center), fontsize = 16, color = :darkgreen)
    
    
    ### bottom panel: action landscape
    
    ax2 = Axis(f[2, 1], xlabel = "real time", ylabel = "imaginary time")

    timags = range(imag(tmin), stop = imag(tmax), length = 100)
    treals = range(real(tmin), stop = real(tmax), length = 100)
    Svals = [S(tr + im*ti) for tr in treals, ti in timags]    
 
    contourf!(ax2, treals, timags, imag.(-Svals), colormap = :grayyellow, extendlow = :white, extendhigh = :white )
    contour!(ax2, treals, timags, imag.(-Svals), levels=21, color=:grey, linewidth=0.85)   
    contour!(ax2, treals, timags, real.(Svals), levels=21, color=:grey, linewidth=0.85)
    scatter!(ax2, reim(saddles)..., color = :black, marker = :x, markersize = 15)

    ### the linesegments of the contour
    datasegments = Observable([ (sim.active ? 
            (Point2f(reim(points[sim.coord[1]].coord)...),
            Point2f(reim(points[sim.coord[2]].coord)...))
            : nothing) 
        for sim in simplices])
    
    datapoints = Observable([ (p.active ? 
            Point2f(reim(p.coord)) : nothing) 
        for p in points])    
    
    linesegments!(ax2, datasegments, color = :green)
#     scatter!(ax2, datapoints, color = :darkgreen) # this clutters too much
    
    text!(ax2, "-Im(S)", space=:relative, position = (0.01, 0.95), align = (:left, :center), fontsize = 16, color = :white)    

    rowsize!(f.layout, 1, Auto(0.5))
    tightlimits!(ax1)
    
    f
    
    ### animate it
    record(f, "integration-domain-flow.mp4", framerate = 4 ) do io
        for Nsteps in 1:50
            p_new, s_new = get_thimble(S, S_drv, real(tmin), real(tmax),
                Nflow = Nsteps,
                flowstepfactor = 1., h_threshold=-500., Δinit = 1., subdividethreshold=0.5)
            
             integrand_real[] = [(sim.active ? 
                    ((real(p_new[sim.coord[1]].coord),
                        real(exp(im*S(p_new[sim.coord[1]].coord))) ),
                    ( real(p_new[sim.coord[2]].coord), 
                        real(exp(im*S(p_new[sim.coord[2]].coord)))   ))
                    : nothing) 
                for sim in s_new]

            integrand_imag[] = [(sim.active ? 
                    ((real(p_new[sim.coord[1]].coord),
                        imag(exp(im*S(p_new[sim.coord[1]].coord))) ),
                    ( real(p_new[sim.coord[2]].coord), 
                        imag(exp(im*S(p_new[sim.coord[2]].coord)))   ))
                    : nothing) 
                for sim in s_new]  
            
            datasegments[] = [ (sim.active ? 
                    (Point2f(reim(p_new[sim.coord[1]].coord)...),
                    Point2f(reim(p_new[sim.coord[2]].coord)...))
                    : nothing) 
                for sim in s_new]

            datapoints[] = [(Point2f(reim(p.coord))) for p in p_new if p.active]            
    
            autolimits!(ax1) # update limits # would be nice to only update only the y axis
            recordframe!(io)cd
        end
    end

end
```
